```python
from tensorflow.keras.models import load_model
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load m hnh
model = load_model("/content/Untitled.h5")

# Th mc cha nh cn nhn din
folder_path = "/content/human_palm_data/FEMALE"

# Kch thc nh u vo
img_size = (64, 64)

# Nhn lp
labels = ['MALE', 'FEMALE']

# Lp qua 5 nh u tin
for idx, file_name in enumerate(os.listdir(folder_path)):
    if file_name.lower().endswith((".jpg", ".jpeg", ".png")):
        if idx >= 5:
            break

        img_path = os.path.join(folder_path, file_name)
        img = cv2.imread(img_path)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img_resized = cv2.resize(img_rgb, img_size)
        img_input = img_resized.astype("float32") / 255
        img_input = img_input.reshape(1, 64*64*3)

        # D on
        prediction = model.predict(img_input)
        confidence = np.max(prediction)
        predicted_label = labels[np.argmax(prediction)]

        # Hin th nh v kt qu
        plt.imshow(img_rgb)
        plt.axis("off")
        plt.title(f"nh: {file_name}\nD on: {predicted_label} ({confidence*100:.2f}%)")
        plt.show()
```
